```python
from src.Twitter_Acess import Acess
from src.Analyses import SentimentalAnalyses
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
#Set The Variables for the search

#Set the path to Model
path = "model.pickle"
#Set the credentials to acess the twetter API in the bellow order 
#[access_token, access_token_secret, consumer_key, consumer_secret]
credentials = []
#Set the words you wamt to search
words = ['java']
#Set the number of twetts
num = 100
```

```python
api = Acess(*credentials)
model = SentimentalAnalyses()
model.load(path=path)
process = api.search(words, count=num)
response = model.predict(process, many=True)
label = ['Positive', 'Nevative']
data = [response.count('Positive'), response.count('Negative')]
index = np.arange(len(label))
plt.bar(index, data)
plt.ylabel('No of Twetts', fontsize=5)
plt.xticks(index, label, fontsize=5, rotation=30)
plt.title('Search for ' + words[0])
plt.show()

```
